Reached on time: Variabel target yang menunjukkan apakah produk tersebut sampai tepat waktu (0) atau tidak (1).

````python
from sklearn import preprocessing

# label_encoder object knows
# how to understand word labels.
label_encoder = preprocessing.LabelEncoder()

df['Warehouse_block'] = label_encoder.fit_transform(df['Warehouse_block'])
df['Mode_of_Shipment'] = label_encoder.fit_transform(df['Mode_of_Shipment'])
df['Product_importance'] = label_encoder.fit_transform(df['Mode_of_Shipment'])
````

````python
X = df[['Warehouse_block', 'Mode_of_Shipment', 'Customer_care_calls',
       'Customer_rating', 'Cost_of_the_Product', 'Prior_purchases',
       'Product_importance', 'Discount_offered', 'Weight_in_gms']]

y = df['Reached.on.Time_Y.N']
````

````python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
````

## Modeling Machine Learning

````python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier()
model.fit(X_train, y_train)
````

````python
print("Akurasi dari model yang sudah di training:")

print("Model DecisionTreeClassifier: ")
print(model.score(X_test, y_test))
````
